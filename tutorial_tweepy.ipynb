```python
import tweepy
```

Kita perlu menggunakan Keys dan Tokens yang diberikan Twitter untuk aplikasi kita. Kunjungi https://developer.twitter.com/ untuk membuat empat buah key rahasia berikut. Setelah itu, isikan ke bagian kode berikut. Anda juga bisa mengikuti step-by-step dari link berikut: https://gist.github.com/MichaelCurrin/da559dcda9f1a0a202b74f2127be5bf9

```python
CONSUMER_KEY = "" # ganti dengan Consumer Key Anda
CONSUMER_SECRET = "" # ganti dengan Consumer Secret Anda
ACCESS_KEY = "" # ganti dengan Access Token Anda
ACCESS_SECRET = "" # ganti dengan Access Token Secret Anda

# WARNING: ketika kalian share kembali kode kalian ke dosen, jangan lupa untuk 
# menghapus 4 secret token tersebut!
```

Uji coba autentikasi dengan Keys dan Tokens kita. Seandainya gagal, kemungkinan besar merupakan masalah jaringan atau ada Keys dan Tokens yang salah.

```python
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_KEY, ACCESS_SECRET)

api = tweepy.API(auth, wait_on_rate_limit=True, wait_on_rate_limit_notify=True)

try:
  api.verify_credentials()
  print("Authentication OK")
except:
  print("Error during authentication")
```

Gunakan contoh 2 dan 3 untuk mengetahui **Mutual Friends** akun Anda! Mutual Friends adalah ketika dimana orang-orang tersebut kita follow & mereka juga follow balik kita.

Hint: kalian mungkin akan memerlukan dua struktur data lists; lalu tampung friends di list pertama; dan tampung followers di list kedua; lalu hitunglah irisannya

```python
screen_name = ''  # Isi dengan username Twitter
search = 50  # Jumlah item(s) yang di search (jangan banyak-banyak)

followers = []
for follower in tweepy.Cursor(api.followers, screen_name).items(search):
    followers.append(follower.screen_name)


following = []
for friend in tweepy.Cursor(api.friends, screen_name).items(search):
    following.append(friend.screen_name)

print(set(followers).intersection(set(following)))
```
